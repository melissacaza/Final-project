```python
import pandas as pd
from vega_datasets import data
import altair as alt
```

```python
dis_df= pd.read_excel('Full_dataset_stripped_normalized.xlsx', sheet_name='dis2')
dis_df.info()
```

```python
dis = dis_df[(dis_df['Seegene_BV_Interpretation'] == 'Bacterial Vaginosis') & (dis_df['BV_reviewed_score_sg'] == 'Negative (0-3)')]
dis.count()
```

```python
dis_lacto = dis[dis['Gene']=='Mobiluncus spp.']
dis_lacto = dis[dis['Gene']=='Lactobacillus spp.']
dis_gv = dis[dis['Gene']=='G. vaginalis']
dis_gv = dis[dis['Gene']=='A. vaginae']
```

```python
lacto = dis_df[(dis_df['Gene'] == 'Lactobacillus spp.') & (dis_df['Agreement2'] == 'Concordant')]
order=['Positive (7-10)','Positive (4-6 with Clue cells)','Negative (4-6 without Clue cells)','Negative (0-3)']

title=alt.TitleParams('Lactobacillus spp.', fontSize=20, fontStyle='Italic')
                                                
lacto_point_bv= alt.Chart(lacto, title=title).mark_point().encode(alt.X('CT:Q', title='Ct value', scale=alt.Scale(domain=[6,38]),axis=alt.Axis(labelAngle=0, labelFontSize=20, titleFontSize=20, grid=True,)),
                                                      alt.Y('BV_reviewed_score_sg:N' ,sort=order, axis=alt.Axis(labels=False,labelFontSize=16, titleFontSize=20, grid=False , labelLimit=900)), 
                                                    alt.Color('BV_reviewed_score_sg:N', legend=None))                                                 


lacto_mean_bv = alt.Chart(lacto).mark_square(size=300).encode(alt.X('mean(CT):Q', title='Ct value',  scale=alt.Scale(domain=[6,38]),axis=alt.Axis(labelAngle=45, labelFontSize=20, titleFontSize=20, grid=True,)),
                                                         alt.Y('BV_reviewed_score_sg:N' ,sort=order, axis=alt.Axis(labels=False,labelFontSize=16, titleFontSize=20, grid=False )),
                                                             alt.Color('BV_reviewed_score_sg:N', legend=None))

error_bars = alt.Chart(lacto).mark_errorbar(extent='stdev',ticks=True, color='black').encode(alt.X('CT:Q', title='Ct value',scale=alt.Scale(domain=[6,38]), axis=alt.Axis(labelAngle=45, labelFontSize=20, titleFontSize=20, grid=True,)),
                                                      alt.Y('BV_reviewed_score_sg:N', sort=order, title= None, axis=alt.Axis(labels=False,labelFontSize=16, titleFontSize=20, grid=False)))

lacto_chart =(lacto_point_bv+ lacto_mean_bv+error_bars ).properties(width=300, height=200).configure_axisY(titleX=-250)
lacto_chart
```

```python
stability_df= pd.read_excel('Seegene Vaginitis analytical performance -stripped.xlsx', sheet_name='Stability')
stability_df.info()
stability_df.head(30)
```

```python
order= ['Lactobacillus spp.', 'Mobiluncus spp.', 'A. vaginae', 'G. vaginalis', 'Candida albicans', 'richomonas vaginalis']
normal_grouped = stability_df[(stability_df['BV_Interpretation'] == 'Normal')& (stability_df['Temperature'] == 22)]
title = alt.TitleParams('BV Negative - 22°C', fontSize=20)
normal_chart= alt.Chart(normal_grouped, title=title).mark_square(size=100).encode(alt.X('Day:O', title='Days', axis=alt.Axis(labelAngle=0, labelFontSize=20, titleFontSize=20, grid=True )),
                                                      alt.Y('mean(Ct value):Q',  scale=alt.Scale(zero=False,domain=[12,35]), title= 'Average Ct value', axis=alt.Axis(labelFontSize=16, titleFontSize=20, grid=True )),
                                                        alt.Color('Organism:N',sort=order, title='Organism'))
                                                            
                                                           
line_normal = alt.Chart(normal_grouped).mark_line().encode(alt.X('Day:O', title='Days', axis=alt.Axis(labelAngle=0, labelFontSize=20, titleFontSize=20, grid=True )),
                                                      alt.Y('mean(Ct value):Q', sort=order, scale=alt.Scale(zero=False,domain=[12,35]), title= None, axis=alt.Axis(labelFontSize=16, titleFontSize=20, grid=True )),
                                                        alt.Color('Organism:N',sort=order, title='Organism'),
                                                          )
                                          
error_bars_norm = alt.Chart(normal_grouped).mark_errorbar(extent='stderr').encode(alt.X('Day:O'),
                                                                          alt.Y('mean(Ct value):Q', title='Average Ct value'),alt.Color('Organism:N', sort=order))
                                                         
                                                  

normal_layer = alt.layer(normal_chart + line_normal + error_bars_norm).properties(width=300, height=400).configure_legend(titleFontSize=20, labelFontSize=18, labelFontStyle='Italic', labelLimit=900)
normal_layer
```
